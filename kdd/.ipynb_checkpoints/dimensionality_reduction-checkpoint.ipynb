# Examples on dimensionality reduction

Sources: Nonlinear Dimensionality Reduction (Lee & Verleysen 2007)

```python
#%matplotlib inline
%matplotlib notebook
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import math
from sklearn.decomposition import PCA
from sklearn.metrics import euclidean_distances

np.random.seed(seed=1337)
seed = np.random.RandomState(seed=1337)
```

## PCA

```python
def embedding_plot_3d(y, **kwargs):
    
    azim = kwargs.get('azim',0)
    elev = kwargs.get('elev',0)
    xlim = kwargs.get('xlim',None)
    ylim = kwargs.get('ylim',None)
    zlim = kwargs.get('zlim',None)
    

    fig = plt.figure(figsize=(9,5))        
    ax = fig.add_subplot(1,1,1)
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    ax.scatter(y.iloc[:, 0], y.iloc[:, 1], y.iloc[:, 2], edgecolor='k', s=40)
    ax.set_title("Embedding in 3D space")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    if xlim != None:
        ax.set_xlim(xlim)
    if ylim != None:
        ax.set_ylim(ylim)
    if zlim != None:
        ax.set_zlim(zlim)
    ax.view_init(azim = azim, elev = elev)
    
```

Create random vector $\mathbf{y}$ by linear transformation of the Gaussian distributed latent variables. In our example latent varible dimension is $\mathbb{R}^{2}$ and the new transformed vector live in $\mathbb{R}^{3}$.

$$\mathbf{y} = \mathbf{W} \mathbf{x}$$

To see how well PCA works, we transform back to $\mathbb{R}^2$ using PCA.

```python
# Create latent variables
x_normal = pd.DataFrame(np.random.randn(1000,2), columns = ['first', 'second'])
x_normal['first'] = 2.5 * x_normal['first'] + 0
x_normal['first'] = 1.5 * x_normal['first'] + 0

# Plot latent variables
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_normal, alpha = 0.5)
ax.set_title('Latent variables in 2D space')

# Linear transformation to 3D space
W = np.matrix('0.2 0.8; 0.4 0.5; 0.7 0.3')
y = pd.DataFrame(np.matmul(x_normal.values,W.T))

embedding_plot_3d(y, azim = 78, elev = 10, xlim=(-4,4),ylim=(-6,5),zlim=(-10,8), presentmode = 'notebook')

# Use PCA to reduce dimensionality back to 2D space
pca_1 = PCA(n_components = 2, random_state = seed)
pca_1.fit(y)
x_pca_1 = pd.DataFrame(pca_1.transform(y))
x_pca_1.columns = ['first','second']
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_normal, alpha = 0.5)
ax.scatter(x = 'first',y = 'second',data = x_pca_1,c = 'purple', alpha = 0.5)
ax.set_title('Original and recovered latent variables in 2D space')
```

Let's create a non-linear transformation and try the same

```python
# Non-linear transformation of latent variables
y = pd.DataFrame()
y['x'] = [4 * math.cos((1/4)*val) for val in x_normal['first']]
y['y'] = [4 * math.sin((1/4)*val) for val in x_normal['first']]
y['z'] = x_normal['first'] + x_normal['second']
embedding_plot_3d(y, azim = 240, elev = 40)

# Use PCA to reduce dimensionality back to 2D space
pca_2 = PCA(n_components = 2)
pca_2.fit(y)
x_pca_2 = pd.DataFrame(pca_2.transform(y))
x_pca_2.columns = ['first','second']
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_normal, alpha = 0.5)
ax.scatter(x = 'first',y = 'second',data = x_pca_2,c = 'purple', alpha = 0.5)
ax.set_title('Original and recovered latent variables in 2D space')
```

Case when latent variables are not Gaussian but rather uniformly distributed:

```python
# Laten variables
a_1 = -8; b_1 = 8
a_2 = -4; b_2 = 4
x_uniform = pd.DataFrame(np.random.rand(1000,2), columns = ['first', 'second'])
x_uniform['first'] = (b_1 - a_1) * x_uniform['first'] + a_1
x_uniform['second'] = (b_2 - a_2) * x_uniform['second'] + a_2
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_uniform, alpha = 0.5)
ax.set_title('Latent variables in 2D space')

# Linear transformation
W = np.matrix('0.2 0.8; 0.4 0.5; 0.7 0.3')
y = pd.DataFrame(np.matmul(x_uniform.values,W.T))
display(y.head())
embedding_plot_3d(y, azim = 140, elev = 20)

# Use PCA to reduce dimensionality back to 2D space
pca_3 = PCA(n_components = 2)
pca_3.fit(y)
x_pca_3 = pd.DataFrame(pca_3.transform(y))
x_pca_3.columns = ['first','second']
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_uniform, alpha = 0.5)
ax.scatter(x = 'first',y = 'second',data = x_pca_3,c = 'purple', alpha = 0.5)
```

```python
from sklearn import manifold

# Plot latent variables
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_normal, alpha = 0.5)
ax.set_title('Latent variables in 2D space')

# Linear transformation to 3D space
W = np.matrix('0.2 0.8; 0.4 0.5; 0.7 0.3')
y = pd.DataFrame(np.matmul(x_normal.values,W.T))
embedding_plot_3d(y, azim = 78, elev = 10, xlim=(-4,4),ylim=(-6,5),zlim=(-10,8))

mds = manifold.MDS(n_components=2,dissimilarity="euclidean", n_jobs=1,random_state=seed)
x_mds = pd.DataFrame(mds.fit_transform(y))
x_mds.columns = ['first','second']

fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_normal, alpha = 0.5, label = 'original')
ax.scatter(x = 'first',y = 'second',data = x_mds,c = 'purple', alpha = 0.5, label = 'mds')
ax.scatter(x = 'first',y = 'second',data = x_pca_1,c = 'forestgreen', alpha = 0.5, label = 'pca')
ax.set_title('Original and recovered latent variables in 2D space')
ax.legend()
```

Scikit learn does not have classical MDS, use another write. Not totally sure how we obtain the Gram matrix here, or I do not see how it is equivalent to 4.20 in Nonlinear Dimensionality Reduction (Lee & Verleysen 2007), but results are very close. Obviously the scikit learn PCA is not the same closed-form version that is equivalent to cMDS, since it is implemented through a probabilistic algortihm. So results cannot be exactly same, but they are very close.

```python
from __future__ import division
 
import numpy as np
 
def cmdscale(D):
    """                                                                                       
    Classical multidimensional scaling (MDS)                                                  
                                                                                               
    Parameters                                                                                
    ----------                                                                                
    D : (n, n) array                                                                          
        Symmetric distance matrix.                                                            
                                                                                               
    Returns                                                                                   
    -------                                                                                   
    Y : (n, p) array                                                                          
        Configuration matrix. Each column represents a dimension. Only the                    
        p dimensions corresponding to positive eigenvalues of B are returned.                 
        Note that each dimension is only determined up to an overall sign,                    
        corresponding to a reflection.                                                        
                                                                                               
    e : (n,) array                                                                            
        Eigenvalues of B.                                                                     
                                                                                               
    """
    # Number of points                                                                        
    n = len(D)
 
    # Centering matrix                                                                        
    H = np.eye(n) - np.ones((n, n))/n
 
    # YY^T                                                                                    
    B = -H.dot(D**2).dot(H)/2
 
    # Diagonalize                                                                             
    evals, evecs = np.linalg.eigh(B)
 
    # Sort by eigenvalue in descending order                                                  
    idx   = np.argsort(evals)[::-1]
    evals = evals[idx]
    evecs = evecs[:,idx]
 
    # Compute the coordinates using positive-eigenvalued components only                      
    w, = np.where(evals > 0)
    L  = np.diag(np.sqrt(evals[w]))
    V  = evecs[:,w]
    Y  = V.dot(L)
 
    return Y, evals
```

```python
Y, evals = cmdscale(euclidean_distances(y))
Y = Y[:,0:2]
Y = pd.DataFrame(Y)
Y.columns = ['first','second']

fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = x_normal, alpha = 0.5, label = 'original')
ax.scatter(x = 'first',y = 'second',data = Y,c = 'purple', alpha = 0.5, label = 'cMDS')
ax.scatter(x = 'first',y = 'second',data = x_pca_1,c = 'forestgreen', alpha = 0.5, label = 'PCA')
ax.set_title('Original and recovered latent variables in 2D space')
ax.legend()
```
