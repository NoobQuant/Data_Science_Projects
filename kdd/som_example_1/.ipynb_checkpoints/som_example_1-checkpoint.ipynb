# Example using SOM
This is a SOM example originally by Bharantendra Rai, found from https://www.youtube.com/watch?v=wLu213JKfnQ. Code here is written based on the video for my own purposes as well as to be suitable for a Jupyter notebook.

```r
# Dependencies
require(kohonen)
require(modeest)
#require(dplyr,lib.loc=.libPaths()[-1]) # force to be loaded from anaconda env lib
require(dplyr)
require(RColorBrewer)
require(repr)
```

```r
# Some colors
coolBlueHotRed = function(n, alpha = 1) {
  rainbow(n, end=4/6, alpha=alpha)[n:1]
}
spectralColors = function(n) {
  brewer.pal(n, "Spectral")[n:1]
}
darkColrors = function(n) {
  #rainbow(n, end=4/6, alpha=alpha)[n:1]
  brewer.pal(n, "Dark2")[1:n]
}
```

- endogenous variable "admit"
- features "gre", "gpa", and "rank". Notice rank is discrete ordinal!

```r
# Get data (fairly small data set)
data_safe = read.csv("binary.csv",header=TRUE)
data = cbind(data_safe)
print(paste0("Frame has ",dim(data)[1]," observations"))
head(data)
```

# Explore full data set with SOM

## Normalize data

```r
# Get all axogenous variables, dropping endogenous ("admit")
# z-score normalization for exogenous variables
X = scale(data[,-1])
head(X)
```

## Train SOM
Suggested grid size: $5*\sqrt{N}$ where $N$ is number of samples.

https://stats.stackexchange.com/questions/282288/som-grid-size-suggested-by-vesanto

```r
5* sqrt(nrow(data))
```

```r
# Train map
# - learning rate alpha: default is to decline linearly from 0.05 to 0.01 over rlen updates
# - radius: the radius of the neighbourhood, either given as a single number or a vector
#           (start, stop). If it is given as a single number the radius will change linearly from
#           radius to zero; a
set.seed(222)
g = somgrid(xdim = 4, ydim = 4, topo = "rectangular")
map = som(X,grid=g,alpha=c(0.05,0.01), radius=1)

set.seed(1)
g2 = somgrid(xdim = 4, ydim = 4, topo = "rectangular")
map2 = som(X,grid=g2,alpha=c(0.05,0.01), radius=1)
```

```r
# Let's plot what the mean distance is, at given iteration, to closest BMU. This basically tells
# how converged the map is.
options(repr.plot.width=7, repr.plot.height=4)
plot(map, type="changes")
```

## Explore density of map

```r
# Create data frame with original data and assigned node on the map
data_extended = cbind(data,map$unit.classif)
names(data_extended) = c(names(data), 'node')

data_extended2 = cbind(data,map2$unit.classif)
names(data_extended2) = c(names(data), 'node')
```

Plot of type "counts" shows the amount of observations mapped into each node. 

```r
# Nodes are numbered from left-to-right, bottom-up
options(repr.plot.width=5, repr.plot.height=5)

plot(map, type="counts",palette.name=coolBlueHotRed)
plot(map2, type="counts",palette.name=coolBlueHotRed)

# Count observations in each node
t(data_extended %>% group_by(node) %>% summarise(count_obs = n()))
t(data_extended2 %>% group_by(node) %>% summarise(count_obs = n()))

print(sum(as.data.frame(data_extended %>% group_by(node) %>% summarise(count_obs = n()))$count_obs))
print(sum(as.data.frame(data_extended2 %>% group_by(node) %>% summarise(count_obs = n()))$count_obs))

# This plots show the points within nodes
plot(map, type="mapping")
plot(map2, type="mapping")
```

Plot type "quality" shows the mean distance of objects mapped to a unit to the codebook vector of that unit.
The smaller the distances, the better the objects are represented by the codebook vectors.

```r
plot(map, type="quality",palette.name=coolBlueHotRed)
plot(map2, type="quality",palette.name=coolBlueHotRed)
```

## Map quality: U-matrix
U-matrix stands for unified distance. Values in U-matrix represent (usually Euclidian) distances of the codebook vectors in input space. Small values mean SOM nodes(codebook vectors) are close together, whereas large values mean SOM nodes are far apart, even if they were close in the output space. As such, the U-matrix can be seen as summary of the probability density function of the input matrix in a 2D space.

The U-matrix should as such have bigger grid than the actual SOM map, since it describes distances between nodes (each distance features two nodes). In the kohonen R package, the grid is the same size. The value in a U-matrix node of the packgae represents the mean (kohonen documentation says **sum** but when inspecting it really looks like mean) of the distances to all immediate neighbours (with unit.distance = 1). Nodes close to each other can be though of as a cluster. This can be a helpful presentation when one tries to find clusters in the input data without having any a priori information about the clusters.

https://stackoverflow.com/questions/13631673/how-do-i-make-a-u-matrix<br>
https://stackoverflow.com/questions/6174011/u-matrix-and-self-organizing-maps

```r
# Distances between 2D SOM grid nodes
unit.distances(g)

# Distances between code vectors in input space
as.matrix(object.distances(map, type = "codes"))

options(repr.plot.width=5, repr.plot.height=5)
plot(map, type="dist.neighbours",palette.name=coolBlueHotRed)
plot(map2, type="dist.neighbours",palette.name=coolBlueHotRed)
```

# Clustering of the SOM map

Approximate optimal k-means cluster amount using "elbow-method". This means we plot the total (sum of) within-cluster sum of squares for k clusters. The location of a bend ("elbow") in the plot is generally considered as an indicator of the appropriate number of clusters. in our case I think the point is k = 4.

https://stats.stackexchange.com/questions/11175/elbow-criteria-to-determine-number-of-cluster<br>
http://www.sthda.com/english/wiki/print.php?id=239#elbow-method

```r
# Get code vectors
codes = getCodes(map)

# For 1 to 15 centers/clusters
wss = c()
for (i in 1:15){
    kmeans = kmeans(codes,centers=i)
    # Within-cluster sum of squares
    crt_wss = kmeans$withinss
    # Sum all together to get total
    wss[i] = sum(crt_wss)    
}

plot(1:15, wss, type="b", xlab="Number of Clusters",ylab="Within groups sum of squares", main="Within cluster sum of squares (WCSS)")
```

```r
# Calculate pairwise distances between codebook vectors
dists = object.distances(map, "codes")
dists2 = object.distances(map2, "codes")

# Hierarchical clustering
m_hclust = hclust(dists)
m_hclust2 = hclust(dists2)

plot(m_hclust)
```

```r
# cutree: Cuts a tree, e.g., as resulting from hclust, into several groups either by specifying the
# desired number(s) of groups or the cut height(s).
clusters = cutree(m_hclust,k=3)
clusters2 = cutree(m_hclust2,k=3)

# cluster for each codebook vector
print(clusters)
print(clusters2)
```

```r
# Colors for clusters
pretty_palette = c("#1f77b4", '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2')

plot(map, type="mapping", bgcol = pretty_palette[clusters], main = "Clusters")
#add.cluster.boundaries(map, clusters) # add cluster lines

plot(map2, type="mapping", bgcol = pretty_palette[clusters2], main = "Clusters")
```

# Using SOM as classifier
SOM canbe used as a classifier, but it's not really the best tool for that...

https://www.quora.com/How-Kohonens-SOM-can-be-used-as-a-classification-tool

```r
set.seed(123)
```

```r
# Create randomized values {1,2} with different frequencies
ind = sample(2, nrow(data), replace=T, prob = c(0.7, 0.3))

# Train and test sets
train = data[ind==1,]
test = data[ind==2,]

# Normalize exogenous train data except admit
trainX = scale(train[,-1])

# Exogenous test set variables, z-scroed with train data specification
testX = scale(test[,-1],
              center = attr(trainX, "scaled:center"),
              scale = attr(trainX, "scaled:scale"))

# Store y's as factors
trainY = factor(train[,1])
testY = factor(test[,1])

# List with test exogenous and a placeholder vector for test predictions.
testXY = list(independent = testX, dependent = rep(0, length(test[,1])))
```

```r
set.seed(222)
```

Function xyf is a wrapper for function "supersom" with two layers. The name stands for X-Y fused SOMs. One vector for each object is created by concatenating X and Y, and a SOM is trained in the usual way, with one exception: the distance of an object to a unit is the sum of separate distances for X and Y spaces. Prediction is done only using the X-space. For continuous Y, the Euclidean distance is used; for categorical Y the Tanimoto distance.

https://www.rdocumentation.org/packages/kohonen/versions/2.0.19/topics/xyf

```r
map_train = xyf(trainX,
           classvec2classmat(trainY), # trainY one-hot encoded
           grid = somgrid(5,5,'hexagonal'),
           rlen = 100)
```

In the plot, black line is X space layer and red for dependent variable layer. ** Don't quite undestand the Y layer!**

```r
options(repr.plot.width=8, repr.plot.height=5)
plot(map_train, type = "changes")
```

```r
t(getCodes(map_train)[[1]])
options(repr.plot.width=5, repr.plot.height=5)
plot(map_train)
```

```r
train_extended = cbind(train,map_train$unit.classif)
names(train_extended) = c(names(train), 'node')

t(train_extended %>% group_by(node) %>% summarise(count_obs = n()))
plot(map_train, type = "count")
```

```r
# Predictions
pred = predict(map_train, newdata = testXY)

# Confusion matrix
table(Predicted = pred$predictions[[2]], Actual = testY)
```

# Get per iteration code vectors for video

```r
iters = 100
g = somgrid(xdim = 3, ydim = 3, topo = "rectangular")
codes = vector("list",iters)
for (l in 1:iters){
    set.seed(222)
    map3 = som(X,grid=g,alpha=c(0.05,0.01), radius=1, rlen = l)
    codes[[l]] = as.data.frame(getCodes(map3))
}

iterno = 10; nodeno = 6
print(codes[[1]]$gre[nodeno])
print(codes[[iterno]]$gre[nodeno])

val = c()
for (i in 1:iters){
    val = c(val, codes[[i]]$gre[nodeno])
}

plot(val)
```

```r
ff = data.frame()
for(i in 1:length(codes)){
    gg = codes[[i]]
    colnames(gg) = c('x','z','y')
    gg[['id']] = paste0("point_",0:(nrow(gg) - 1))
    gg[['iter']] = i
    ff = rbind(ff, gg)
}
write.csv(ff, file = "codes_all.csv", row.names=F)    
```
