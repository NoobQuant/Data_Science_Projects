```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(seed=1337)
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns = ['first', 'second'])
df['first'] = 2.5 * df['first'] + 0
df['first'] = 1.5 * df['first'] + 0

fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = df, alpha = 0.5)
```

```python
W = np.matrix('0.2 0.8; 0.4 0.5; 0.7 0.3')
M = df[['first','second']].values
M_new = pd.DataFrame(np.matmul(M,W.T))
M_new.head()

fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(1,1,1)
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.scatter(M_new.loc[:, 0], M_new.loc[:, 1], M_new.loc[:, 2], edgecolor='k', s=40)
ax.set_title("Embedded in 3D space")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_xlim(-4,4)
ax.set_ylim(-6,5)
ax.set_zlim(-10,8)
ax.view_init(azim = 35, elev = 10)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(M_new)
M_transformed = pd.DataFrame(pca.transform(M_new))
M_transformed.columns = ['first','second']

fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(1,1,1)
ax.scatter(x = 'first',y = 'second',data = df, alpha = 0.5)
ax.scatter(x = 'first',y = 'second',data = M_transformed,c = 'purple', alpha = 0.5)
```
